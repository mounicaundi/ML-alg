```python
#converting the categorical into numerical
dt = pd.get_dummies(dt,drop_first=True) ## dropfirst will stop splitting and droping one of the col(basic)
dt
```

```python
dt['category_Luxury'].value_counts() ##  basic are 7470 whereas lux-2350
```

```python
#splitting x and y
X = dt.drop('category_Luxury', axis=1)
y = dt['category_Luxury']
```

```python
y.shape
```

```python
#splitting in training and testing values
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 0,test_size = 0.3)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#model buliding
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
#predictions
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
con_mat = confusion_matrix(y_test,y_pred)
con_mat
```

```python
accuracy_score(y_test,y_pred)
```
