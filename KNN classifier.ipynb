```python
#droping User ID column as it is not an important feature
df.drop('id',axis =1,inplace = True) ## axis is the postion of the coloumn , inplace is the operation
```

```python
df.head()
```

```python
#converting categorical values to numerical
df = pd.get_dummies(df,drop_first= True)
```

```python
df.head()
```

```python
#seperating X and y
X = df.loc[:,['Age','Driving_License','Region_Code','Previously_Insured','Annual_Premium',
              'Policy_Sales_Channel','Vintage','Gender_Male','Vehicle_Age_< 1 Year','Vehicle_Age_> 2 Years',
              'Vehicle_Damage_Yes']].values
y = df.loc[:,['Response']].values
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

```python
y.shape
```

```python
#model was expecting a 1d array as input
y = y.reshape(-1)
```

```python
y.shape
```

```python
#Splitting the data into train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 0, test_size = 0.25)
```

```python
#importing classification metrics
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
acc_list = []
err_list = []
acc_list = []
err_list = []
for i in range(1,5):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test,y_pred)
#     print("For k = {}, accuracy = {}".format(i,acc))
    acc_list.append(acc)
    err_list.append(1-acc)
```

```python
plt.plot(list(range(1,5)),err_list,c= 'r')
plt.title('Error rate v/s K')
plt.xlabel('K')
plt.ylabel('Error rate')
plt.show()
```
